```
import numpy as np
from PIL import Image, ImageOps
from matplotlib import pyplot as plt
np.set_printoptions(suppress=True)
A=np.random.randint(low=10,size=(3,4))
print(A)
```

```
U, S, V = np.linalg.svd(A)
print(U)
print("-"*50)
print(S)
print("-"*50)
print(V)
signma = np.zeros((3,4))
print(signma)
for i, x in enumerate(S):
  signma[i, i] = x
print(signma)
```

```
imag = Image.open('/content/ngan.jpg')
imag.show('/content/ngan.jpg')
imag.save('/content/ngan.jpg')
plt.imshow(imag)
imag = ImageOps.grayscale(imag)
plt.imshow(imag, cmap = 'autumn_r')
#print(ma)
```

```
U, S, V = np.linalg.svd(imag)
plt.plot(S, 's-')
#plt.xlim([0,300])
#plt.ylim([500,3000])
plt.xlabel('s thnh phn')
plt.ylabel('Gi tr suy bin')
plt.show()
```

```
comps = np.arange(0, 50)
#comps = np.arrange(20,100)
rec_pic = U[:,comps]@np.diag(S[comps])@V[comps, :]
plt.figure(figsize = (10,200))
plt.imshow(imag, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(rec_pic, cmap='gray_r')
plt.title(f'reconstructed pic from {comps[0]} to {comps[-1]}')
plt.axis('off')

plt.show
```

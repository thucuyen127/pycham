4. Cách lấy các url trong html

```
from bs4 import BeautifulSoup
import requests

url = 'http://vietnamnet.vn//'
depth = 3
count = 3
history = [[] for i in range(0, depth + 1)]
history[0].append(url)
for depth_i in range(0, depth):
    for links in history[depth_i]:
        response = requests.get(links)
        soup = BeautifulSoup(response.text, 'html.parser')
        tags = soup.find_all('a')
        for link in tags:
            url_new = link.get('href')
            flag = False
            for item in history:
                for l in item:
                    if url_new == l:
                        flag = True

            if url_new is not None and "https" in url_new and flag is False:
                history[depth_i + 1].append(url_new)
                print(url_new)
```
